### Remplzar los signos de interrogación por NaN

```python
df.replace("?", np.nan, inplace=True)
```

### Convert column titles "city-mpg" and "highway-mpg" to "city-kpl", "highway-kpl". Maybe
### you need to add new columns, investigate the command.
### Realizar la conversión


```python
kpl_factor = 0.425144 #Factor de conversión de millas por galón a kilómetros por litro
# Crear las nuevas columnas con el cálculo:
df["city-kpl"] = df["city-mpg"] * kpl_factor
df["highway-kpl"] = df["highway-mpg"] * kpl_factor
# Validar el resultado
print(df[["city-kpl", "highway-kpl"]].head())
print(df[["city-mpg","city-kpl", "highway-mpg","highway-kpl"]].head())
```

### Limpieza de datos



```python
# Limpieza rápida de la columna precio
df['price'] = pd.to_numeric(df['price'], errors='coerce')
# Eliminar filas sin precio
df.dropna(subset=['price'], inplace=True)
```

### Generar un gráfico


```python
# Gráfica de cajas (Boxplot) para ver distribución por marcas
plt.figure(figsize=(12, 6)) #Reservar el espacio para la gráfica
sns.boxplot(x='price', y='make', data=df) #Gráfica de cajas (Boxplot) para ver distribución por marcas
plt.title('Distribución de Precios por Marca') #Título de la gráfica
plt.show()
```

### Gráfico de calor


```python
# Seleccionamos variables numéricas relevantes
corr_matrix = df[['engine-size', 'horsepower', 'curb-weight', 'city-kpl', 'price']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='Spectral', fmt=".2f")
plt.title('Mapa de Calor: ¿Qué influye en el precio y el consumo?')
plt.show()
```

### Gráfica de regresión

```python
# Gráfica: Scatter Plot con Línea de Regresión
plt.figure(figsize=(10, 6))
sns.regplot(x='engine-size', y='city-kpl', data=df, color='darkblue')
plt.title('Relación: Tamaño del Motor vs. Kilómetros por Litro')
plt.xlabel('Tamaño del Motor (Engine Size)')
plt.ylabel('Eficiencia (City KPL)')
plt.show()
```

### Agrupación o Bining

```python
# Crear 3 grupos (bins) de precios
# Grupo 1: Desde el precio más bajo (df['price'].min()) hasta 15,000.
# Grupo 2: Desde 15,001 hasta 30,000.
# Grupo 3: Desde 30,001 hasta el precio más alto (df['price'].max())
bins = [df['price'].min(), 15000, 30000, df['price'].max()]
nombres = ['Económico', 'Estándar', 'Lujo']
df['price-category'] = pd.cut(df['price'], bins, labels=nombres, include_lowest=True)
# Visualizar cuántos hay de cada uno
sns.countplot(x='price-category', data=df, palette='Set3')
plt.title('Distribución de Autos por Categoría de Precio')
plt.show()
```
